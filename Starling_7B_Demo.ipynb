```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
```

```python
tokenizer = AutoTokenizer.from_pretrained("berkeley-nest/Starling-LM-7B-alpha")
model = AutoModelForCausalLM.from_pretrained("berkeley-nest/Starling-LM-7B-alpha", torch_dtype = torch.float16, device_map="auto")
```

## Text Gen Function

```python
def generate_response(prompt):
  input_ids = tokenizer(prompt, return_tensors="pt").input_ids
  outputs = model.generate(input_ids.to(model.device), temperature = 0.3, max_new_tokens=1024, do_sample=True, pad_token_id = tokenizer.pad_token_id, eos_token_id = tokenizer.eos_token_id)
  return tokenizer.decode(outputs[0][input_ids.size(1) :])
```

## Prompt Format
OpenChat Format

### GPT4 Correct User: {prompt} <|end_of_turn|>GPT4 Correct Assistant:

```python
prompt = "What is AI?"
```

```python
single_turn_prompt = f"GPT4 Correct User: {prompt}<|end_of_turn|>GPT4 Correct Assistant:"
```

```python
response_text = generate_response(single_turn_prompt)
```

```python
print("Response:", response_text)
```

```python
multi_turn_prompt = "What is AI?"
follow_up_ques = "What are the pros and cons of AI?"
response = " AI, or Artificial Intelligence, refers to the development of computer systems or software that can perform tasks that typically require human intelligence."

multi_turn_prompt = f"GPT4 Correct User: {prompt}<|end_of_turn|>GPT4 Correct Assistant: {response}<|end_of_turn|>GPT4 Correct User: {follow_up_ques}<|end_of_turn|>GPT4 Correct Assistant:"
```

```python
multi_turn_response = generate_response(multi_turn_prompt)
```

```python
print("Multi-turn conversation response:", multi_turn_response)
```

```python
code_prompt = "Write a FastAPI code to develop a REST API."
coding_prompt = f"Code User: {code_prompt}<|end_of_turn|>Code Assistant:"
```

```python
response_code = generate_response(coding_prompt)
```

```python
print("Coding Conversation Response:", response_code)
```
